```python
import numpy as np
from scipy.integrate import odeint
from matplotlib import pyplot as plt
from scipy.special import kn, zeta
from scipy.interpolate import interp1d
```

```python
# Get process dictionary
sigmaVfile = './test_Coann/output/taacs.csv'
with open(sigmaVfile) as f:
    comment_lines =  [l.strip() for l in f.readlines() if l.strip()[0] == '#']
    proc_lines = [l[1:].strip() for l in comment_lines if (l.count('#') == 1)]
processDict = {}
for l in proc_lines:    
    proc_index,proc_name,proc_pdgs = (l.split(',',2))
    proc_pdgs = proc_pdgs.replace('[','').replace(']','')
    initialPDGs,finalPDGs = proc_pdgs.split('_')
    processDict[proc_index] = {'name' : proc_name.strip(), 
                            'initialPDGs' : list(map(int, initialPDGs.split(','))), 
                            'finalPDGs' : list(map(int, finalPDGs.split(',')))}
processes_data = np.genfromtxt(sigmaVfile, delimiter=',', skip_header=len(comment_lines), 
                            names=True)
xlist = processes_data['x']
sV = np.zeros(len(xlist))
for iproc in processDict:
    if iproc == 34: continue
    sV = sV + processes_data[str(iproc)]
sigmaV = interp1d(xlist,sV,fill_value="extrapolate",bounds_error=False)
```

```python
from sympy.parsing.mathematica import parse_mathematica
from sympy import var
x = var('x')
expr = "-(F[x]*G[x]^2) - (2*F[x]^2*G[x]*Yeq[x]^2)/(1 + F[x]*Yeq[x]) + (F[x]*Yeq[x]^2*(1 + 2*F[x]*Yeq[x]))/(1 + F[x]*Yeq[x])^2 -(Yeq[x]^2*Derivative[1][F][x])/(1 + F[x]*Yeq[x])^2 + (-1 + (1 + F[x]*Yeq[x])^(-2))*Derivative[1][Yeq][x]"
expr = expr.replace('Derivative[1][F][x]','dFdx[x]').replace('Derivative[1][Yeq][x]','dYeqdx[x]')
python_code = parse_mathematica(expr)
# print(python_code)
python_simp = str(python_code).replace('F(x)','f').replace('G(x)','g').replace("Yeq(x)","yeq").replace("/(f*yeq**2 + 1)","*t0")
print(python_simp)
```

```python
def Yeq(x):
    #m represents particle's mass, T is temperature in GeV, g represents the particle's degrees of freedom, 
    #and s represents entropy density.
    m = 1.0
    mDM = 1.0
    g = 1.0
    T = mDM/x
    xm = (m/mDM)*x

    g_star = 1.0 #g_* is the number of relativistic degrees of freedom
    s = (2.*np.pi**2/45.)*g_star*m**3/xm**3 #entropy density
    if xm > 10: #non-relativistic regime
        coeffs = [1,15./8.,105./128.,-315/1024.,10395./32768.]
        neq = (m**3*np.exp(-xm)/(xm**(3/2)))*(1/(2*np.sqrt(2)*np.pi**(3/2)))
        neq = neq*sum([c/xm**i for i,c in enumerate(coeffs)])
    elif xm > (2./3.): #semi-relativistic regime
        neq = m**3*(1/(2*np.pi**2))*kn(2, xm)/xm
    else: #relativistic regime
        neq = (zeta(3)/np.pi**2)*T**3 # bosons
        if g > 0:
           neq = (3/4)*neq # fermions
            
    Y_eq = abs(g)*neq/s
    
    return Y_eq


def sigmaVFunc(x):

    return sigmaV(x)
    

def F(x):

    return 1e10*sigmaVFunc(x)/x**2


def dYeqdx(x):

    dx = 1e-4
    return (Yeq(x+dx)-Yeq(x))/dx

def dFdx(x,ep=1e-2):

    fp = F(x+ep)
    fm = F(x-ep)
    df = (F(x+ep)-F(x-ep))/(2*ep)

    return df

def YtoG(Y,x):
    yeq = Yeq(x)
    f = F(x)
    fy = f*yeq
    t1 = fy/(1+fy)
    return Y - yeq*t1

def GtoY(G,x):
    yeq = Yeq(x)
    f = F(x)
    fy = f*yeq
    return G + yeq*(1-1/(1+fy))

def dYdx(Y,x):

    f = F(x)
    
    return f*(Yeq(x)**2 - Y**2)
```

```python
def dGdx(G,x):

    f = F(x)
    g = G
    yeq = Yeq(x)
    fy = f*yeq
    fg = f*G
    
    t0 = 1/(1+fy)
    
    # d = -fg*g - 2*fy*fg*yeq*t0 - (fy + 3)*fy*dYeqdx(x)*t0**2 + (2*fy + 1)*fy*t0**2 - yeq**3*dFdx(x)*t0**2

    d = (-1 + (fy + 1)**(-2))*dYeqdx(x) - f*g**2 - 2*fy**2*g/(fy + 1) + (2*fy + 1)*fy*yeq/(fy + 1)**2 - yeq**2*dFdx(x)/(fy + 1)**2




    return d
```

```python
x0 = 0.1
Y0 = Yeq(x0)
G0 = YtoG(Y0,x0)
xvals = np.linspace(x0,100.,1000)

print(Yeq(x0),F(x0),dYeqdx(x0),dFdx(x0),G0,dGdx(G0,x0))
```

```python
G_sol,gDict = odeint(dGdx,G0,xvals,full_output=True)
Y_sol0,yDict = odeint(dYdx,Y0,xvals,full_output=True)
```

```python
Y_sol = [GtoY(G,x) for G,x in zip(G_sol,xvals)]
Y_eq = [Yeq(x) for x in xvals]
dGdx_sol = [dGdx(G,x) for G,x in zip(G_sol,xvals)]
dYdx_sol = [dYdx(Y,x) for Y,x in zip(Y_sol0,xvals)]
F_sol = [F(x) for x in xvals]
dFdx_sol = [dFdx(x) for x in xvals]
dYeqdx_sol = [dYeqdx(x) for x in xvals]
```

```python
plt.plot(xvals,Y_sol,label='Y from dGdx')
plt.plot(xvals,Y_sol0,label='Y from dYdx')
plt.plot(xvals,Y_eq,linestyle='dashed',label='Yeq')
plt.plot(xvals,dGdx_sol,linestyle='dotted',label='dGdx')
plt.plot(xvals,dYdx_sol,linestyle='dotted',label='dYdx')

plt.legend()
plt.yscale('log')
plt.xscale('log')
# plt.ylim(1e-10,1e-1)
plt.ylim(1e-4,10)
plt.grid()
plt.title(r'$Y_{f} = $%1.3e (from dYdx), %1.3e (from dGdx)' %(Y_sol0[-1][0],Y_sol[-1][0]))
plt.show()
```
